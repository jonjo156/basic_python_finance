```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
K = 3000 #option strike
```

```python
S = np.linspace(2000, 5000, 200) #inner values of call option
```

```python
h = np.maximum(S-K, 0) #inner value of call option
```

```python
plt.figure()
plt.plot(S, h, lw=2.5)
plt.xlabel('index level $S_t$ at maturity')
plt.ylabel('inner value of S&P 500 call option')
plt.grid(True)
```
